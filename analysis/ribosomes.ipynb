```python
# Radial distribution (probability density vs r/R) — same as plot_radial_distribution.py
import matplotlib.pyplot as plt
from plot_radial_distribution import gather_files, ensemble_average_density

def plot_ribo_distribution(date, minute, rep, title, legend):
    # Parameters: date + replicate/timepoint ranges; files in {base_dir}/{date}_{rep}/data/coords/
    R = 2500.0
    base_dir = "../runs/"  # project root containing Jan30_1/, Jan30_2/, ...
    date = date
    rep_start = rep
    tp_start = minute  # timepoints (0-based in filenames)
    order = "row"
    rep_range = range(rep_start, rep_start + 1)
    tp_range = range(tp_start, tp_start + 1)
    dna_files = gather_files("dna", date, rep_range, tp_range, base_dir, replicate_subdirs=True)
    ribo_files = gather_files("ribo", date, rep_range, tp_range, base_dir, replicate_subdirs=True)
    if not dna_files: print("Warning: no DNA files found.")
    if not ribo_files: print("Warning: no ribosome files found.")
    centers, dna_density = ensemble_average_density(dna_files, R, order=order)
    _, ribo_density = ensemble_average_density(ribo_files, R, order=order)
    fig, ax = plt.subplots(figsize=(6, 4))
    width = centers[1] - centers[0]
    ax.bar(centers, dna_density, width=width, color="limegreen", alpha=0.5, label="DNA")
    ax.bar(centers, ribo_density, width=width, color="magenta", alpha=0.5, label="Ribosomes")
    ax.set_xlabel(r"Normalized radius $r/R$")
    ax.set_ylabel(r"$p(r) \times R^3$")
    ax.set_title(title)
    if legend:
        ax.legend()
    ax.tick_params(axis="both", labelsize=14)
    ax.set_xlabel(ax.get_xlabel(), fontsize=14)
    ax.set_ylabel(ax.get_ylabel(), fontsize=14)
    ax.set_title(ax.get_title(), fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
minute = 0
plot_ribo_distribution("Mar03", minute, 5, "20/350/75, 0 min", True)
#plot_ribo_distribution("Mar03", minute, 4, "0.64 (20/350/50)", False)
plot_ribo_distribution("Mar03", minute, 18, "35/500/100, 0 min", False)
#plot_ribo_distribution("Mar03", minute, 25, "1.84 (50/200/100)", False)
```
